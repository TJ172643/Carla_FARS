```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("./data/FARS2021NationalCSV/vehicle.csv", encoding="windows-1252")
```

## Blind spot crash filter

```python
bs_crash = df[(df['IMPACT1']<=9) & (df['IMPACT1']>=7)|(df['IMPACT1']<=5) & (df['IMPACT1']>=3)|(df['IMPACT1']==63)|(df['IMPACT1']==83)]
```

## Find case number, involved vehicle number less than 3

```python
bs_case_list = set(bs_crash[bs_crash['VE_FORMS']<3]['ST_CASE'].unique())
print(len(bs_case_list))
```

```python
unknow_speed_list = set(df[df['TRAV_SP']>997]['ST_CASE'].unique())
print(len(unknow_speed_list))
```

```python
case_list = bs_case_list - unknow_speed_list
print(len(case_list))
```

```python
# get rows with case_list
test = df[df['ST_CASE'].isin(case_list)]
test['P_CRASH1'].value_counts()
```

```python
veh1 = set(test[test['P_CRASH1']==1]['ST_CASE'])
veh2 = set(test[test['P_CRASH1']==11]['ST_CASE'])
case_1_11_list = veh1.intersection(veh2)
print(case_1_11_list)
```
